```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read School and Student Data File and store into Pandas Data Frames
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
mergedata_df = pd.merge(ride_data, city_data, how="left", on=["city"])
mergedata_df.head()
```

## Bubble Plot of Ride Sharing Data

```python

#group by city types
city_types = mergedata_df.groupby('type')

#define for each type of the three city
urban = city_types.get_group('Urban')
rural = city_types.get_group('Rural')
suburban =city_types.get_group('Suburban')

#define variable to calculate the total and average of the each city type
urban_total_rides = urban.groupby('city')['ride_id'].count()
urban_average_fare = urban.groupby('city')['fare'].mean()
urban_total_drivers= urban.groupby('city')['driver_count'].sum()

rural_total_rides = rural.groupby('city')['ride_id'].count()
rural_average_fare = rural.groupby('city')['fare'].mean()
rural_total_drivers= rural.groupby('city')['driver_count'].sum()

suburban_total_rides = suburban.groupby('city')['ride_id'].count()
suburban_average_fare = suburban.groupby('city')['fare'].mean()
suburban_total_drivers= suburban.groupby('city')['driver_count'].sum()

plt.scatter(urban_total_rides,urban_average_fare,urban_total_drivers, c='r', alpha=0.2, label="Urban")
plt.scatter(rural_total_rides,rural_average_fare,rural_total_drivers, c='y', alpha=0.7, label="Rural")
plt.scatter(suburban_total_rides,suburban_average_fare, suburban_total_drivers, c='b',alpha=0.3, label="Suburban")
plt.title('Pyber Ride Sharing Data 2016')
#plt.set_xlabel("Total number of rides per city")
#plt.set_ylabel("Average Fare")
plt.legend()
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python

#group by total fare 
#city_types = mergedata_df.groupby('type')
Total_city = mergedata_df.groupby('city')['fare'].count()
#Total_city

#define variable to calculate the total of the each city type
urban_fare= urban.groupby('city')['fare'].count()
rural_fare = rural.groupby('city')['fare'].count()
suburban_fare = suburban.groupby('city')['fare'].count()
#suburban_fare

#calculate percentage city
urban_percentage = len(urban_fare)/len(Total_city)
rural_percentage = len(rural_fare)/len(Total_city)
suburban_percentage = len(suburban_fare)/len(Total_city)
#suburban_percentage

#Ploting pie chart
# define, labels ,colors, fracs 
labels = ['Urban', 'Rural', 'Suburban']
colors = ['pink', 'yellow', 'skyblue']
explode = (0.05, 0, 0)
fracs = [urban_percentage,rural_percentage,suburban_percentage]
plt.pie(fracs, explode=explode, labels=labels,colors=colors,autopct='%1.1f%%', shadow=True, startangle=250)
plt.title('% of Total Fares by City Type')
#plt.title('% of Total Fares by City Type', bbox={'facecolor':'0.7', 'pad':5})

plt.show()
```
